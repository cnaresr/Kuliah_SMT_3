```python
import numpy as np
import matplotlib.pyplot as plt

# --- Sinyal Kontinu (asli) ---
t_cont = np.linspace(0, 1, 1000)   # waktu 0 - 1 detik (1000 titik)
x_cont = np.sin(2 * np.pi * 5 * t_cont)  # sinyal sinus 5 Hz

# --- Sampling dari sinyal kontinu (diskrit) ---
fs = 50  # frekuensi sampling 20 Hz
t_samp = np.arange(0, 1, 1/fs)   # waktu sampling
x_samp = np.sin(2 * np.pi * 5 * t_samp)  # nilai sinyal pada titik sampling

# --- Plot Sinyal Kontinu ---
plt.figure(figsize=(8, 4))
plt.plot(t_cont, x_cont, label='Sinyal Kontinu (Asli)', color='b')
plt.title("Sinyal Kontinu (Continuous Signal)")
plt.xlabel("Waktu (detik)")
plt.ylabel("Amplitudo")
plt.legend()
plt.grid(True)
plt.show()

# --- Plot Sinyal Diskrit (Hasil Sampling) ---
plt.figure(figsize=(8, 4))
plt.stem(t_samp, x_samp, linefmt='r-', markerfmt='ro', basefmt=" ")
plt.title("Sinyal Diskrit (Sampling dari Kontinu)")
plt.xlabel("Waktu (detik)")
plt.ylabel("Amplitudo")
plt.grid(True)
plt.show()
```
